```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
from imblearn.ensemble import BalancedRandomForestClassifier
from imblearn.pipeline import Pipeline
from sklearn.metrics import roc_curve, auc, classification_report
from sklearn.metrics import classification_report
```

```python
# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:, :-1], data.iloc[:, -1], test_size=0.3, random_state=42)
```

```python
import warnings

warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
# Define the models
models = [
    ('Logistic Regression', LogisticRegression(class_weight='balanced', max_iter=1000)),    
    ('Decision Tree', DecisionTreeClassifier(class_weight='balanced')),    
    ('RUS + LogReg', Pipeline([('rus', RandomUnderSampler()), ('log', LogisticRegression(class_weight='balanced', max_iter=1000))])),
    ('ROS + LogReg', Pipeline([('ros', RandomOverSampler()), ('log', LogisticRegression(class_weight='balanced', max_iter=1000))])),
    ('SMOTE + LogReg', Pipeline([('smote', SMOTE()), ('log', LogisticRegression(class_weight='balanced', max_iter=1000))])),
    ('Balanced Random Forest', BalancedRandomForestClassifier())
]
```

```python
# Fit and evaluate each model
results = []
names = []
for name, model in models:
    if isinstance(model, Pipeline):
        model.fit(X_train, y_train)
        transformer = model.named_steps['rus'] if 'rus' in model.named_steps else model.named_steps['ros'] if 'ros' in model.named_steps else model.named_steps['smote']
        X_resampled, y_resampled = transformer.fit_resample(X_train, y_train)
        y_pred = model.named_steps['log'].predict(X_test)
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        X_resampled, y_resampled = X_train, y_train
    fpr, tpr, _ = roc_curve(y_test, model.predict_proba(X_test)[:,1])
    roc_auc = auc(fpr, tpr)
    report = classification_report(y_test, y_pred, output_dict=True)
    results.append([roc_auc, report['0']['recall'], report['1']['recall']])
    names.append(name)
    
    # Print confusion matrix and classification report
    print(name)
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
```

```python
fig, ax = plt.subplots(figsize=(12,6))
bar_width = 0.25
opacity = 0.8
index = np.arange(len(names))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
for i in range(len(results[0])):
    ax.bar(index + i * bar_width, [r[i] for r in results], bar_width, alpha=opacity, color=colors[i], label=['AUC', 'Non-fraudulent Recall', 'Fraudulent Recall'][i])
    # Add the values on top of each bar
    for j, r in enumerate(results):
        ax.text(index[j] + i * bar_width, r[i]+0.01, f"{r[i]:.2f}", ha='center', va='bottom', fontweight='bold')
        
ax.set_xlabel('Model', fontsize=12)
ax.set_ylabel('Score', fontsize=12)
ax.set_title('Comparison of different models', fontsize=14)
ax.set_xticks(index + bar_width)
ax.set_xticklabels(names)  # Rotate x-axis labels by 30 degrees and align right
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3, fontsize=12)  # Set legend location and font size

plt.tight_layout()
plt.show()
```
